```python
class Node:
    def __init__(self, node_id, obj=None):
        self.par = []
        self.chids = []
        self.name = node_id
    
def get_root(node):
    if not node.par:
        return [node]
    all_parents = []
    for par in node.par:
        all_parents.extend(get_root(par))
    return all_parents
```

```python
root = Node('Root', obj="Hi this is root")
```

```python
n1 = Node('node1', obj="Node 1")
n1.par.append(root)
```

```python
n2 = Node('node2', obj="Node 2")
n2.par.append(root)
```

```python
n2_2 = Node('node2_2', obj="Node 2 ka baccha")
n2_2.par.append(n2)
```
